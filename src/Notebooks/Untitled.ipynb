```python
import pandas as pd

data = {'Ambito Geografico': ['Municipio', 'Municipio', 'Municipio', 'Municipio',
                             'Municipio', 'Municipio', 'Municipio', 'Municipio',
                             'Municipio', 'Municipio', 'Municipio', 'Municipio',
                             'Municipio', 'Municipio', 'Municipio', 'Municipio'],
        'Regiao': ['Norte', 'Norte', 'Norte', 'Norte', 'Centro', 'Centro', 'Centro',
                   'Centro', 'Area Metropolitana de Lisboa', 'Area Metropolitana de Lisboa',
                   'Area Metropolitana de Lisboa', 'Area Metropolitana de Lisboa',
                   'Area Metropolitana de Lisboa', 'Area Metropolitana de Lisboa',
                   'Area Metropolitana de Lisboa', 'Area Metropolitana de Lisboa'],
        'Cidades': ['Arcos de Valdevez', 'Caminha', 'Melgaco', 'Moncao', 'Lourinha',
                    'Nazare', 'Obidos', 'Peniche', 'Amadora', 'Barreiro', 'Cascais',
                    'Lisboa', 'Loures', 'Mafra', 'Moita', 'Montijo'],
        'qtd Hospitais particulares e publicos': [0, 0, 0, 0, 1, 2, 3, 10, 3, 1, 6, 33, 2, 0, 0, 1],
        'qtd Farmacias': [6, 5, 3, 6, 6, 5, 3, 6, 41, 23, 42, 256, 53, 15, 16, 14],
        'Quantidade construcoes novas': [20, 20, 1, 10, 20, 20, 1, 10, 7, 21, 112, 10, 96, 175, 32, 63]
       }

df = pd.DataFrame(data)


```

```python
df['SomaTudo'] = df['qtd Hospitais particulares e publicos'] + df['qtd Farmacias'] + df['Quantidade construcoes novas']

df
```

```python
df['ranking Geral'] = (df['SomaTudo'] / df['SomaTudo'].sum()) * 100

df
```

```python
df['ranking Regional'] = df.groupby('Regiao')['SomaTudo'].apply(lambda x: x / x.sum() * 100)
```

```python
# lista de regiões
regioes = ['Centro', 'Norte']

# criação do novo dataframe com as colunas ranking Geral e ranking Regional
novo_df = pd.DataFrame({'Ambito Geografico': df['Ambito Geografico'], 'Regiao': df['Regiao'], 'Cidades': df['Cidades'], 'ranking Geral': 0, 'ranking Regional': 0})

# laço for para percorrer as regiões
for regiao in regioes:
    # criação do dataframe auxiliar com os dados filtrados pela região atual do laço
    aux_df = novo_df[novo_df['Regiao'] == regiao]
    
    # cálculo da soma total da coluna SomaTudo para a região atual
    soma_regiao = aux_df['SomaTudo'].sum()
    
    # atualização do ranking Regional para cada cidade da região atual
    for index, row in aux_df.iterrows():
        novo_df.loc[index, 'ranking Regional'] = row['SomaTudo'] / soma_regiao * 100
        
# atualização do ranking Geral para cada cidade do dataframe completo
soma_total = novo_df['SomaTudo'].sum()
for index, row in novo_df.iterrows():
    novo_df.loc[index, 'ranking Geral'] = row['SomaTudo'] / soma_total * 100

# exibição do novo dataframe
novo_df
```
