```python
df1['Ranking Geral'] = df1['Clientes'].rank(method='dense', ascending=False)
df1 = df1.sort_values(by=['Ranking Geral'], ascending=True)
```

```python
df1['Ranking por Regiao'] = df1.groupby('Regiao')['Clientes'].rank(method='dense', ascending=False)
df1 = df1.sort_values(by=['Regiao', 'Ranking por Regiao'], ascending=[True, True])
```
