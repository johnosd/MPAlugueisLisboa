```python
from Geocoding.Geocoding import google_maps, geo_py
import pandas as pd
```

```python
gmaps = google_maps()
geopy = geo_py()
```

```python
dir_path ='..\\..\\Bases\\Municipios\\'
```

```python
file_municipios = 'Municipios.csv'

file_path = '{}{}'.format(dir_path,file_municipios)
df_municipios = pd.read_csv(file_path, sep=";", decimal=',')
```

```python
# Geocodifica municipios
geopy.geopy_geocode(df_municipios)
```

```python
# Cria coluna de distancia das principais cidades de portugal
cidades =['lisboa, Portugal','porto, Portugal']
geopy.geopy_distance(df_municipios,cidades)
```
