```python
from Geocoding.Geocoding import google_maps, geo_py
import pandas as pd
```

```python
gmaps = google_maps()
geopy = geo_py()
```

```python
dir_path ='..\\..\\Bases\\'
```

```python
file_municipios = 'Municipios.csv'

file_path = '{}{}'.format(dir_path,file_municipios)
df_municipios = pd.read_csv(file_path, sep=";", decimal=',')
```

```python
print('-- GEO_PY - Geocoding Started -----------------')
geopy.geopy_geocode(df_municipios)
    
print('-- GMAPS - Geocoding Started ------------------')
gmaps.googlemaps_city_distance_from_city(df_municipios)
    
print('-- Geocoding Ended ----------------------------')
```

```python
df_municipios.rename({'duration_hours_driving_lisboa, Portugal':'time_driving_lisboa',
                      'distance_drivingdriving_lisboa, Portugal':'distance_driving_lisboa',
                      'duration_hours_driving_porto, Portugal':'time_driving_porto',
                      'distance_drivingdriving_porto, Portugal':'distance_driving_porto'},inplace=True)
```

```python
file_municipios_geo = 'Municipios_Geo.csv'
file_path = '{}{}'.format(dir_path,file_municipios_geo)
df_municipios.to_csv(file_path, sep=';', quotechar='"', decimal=',')
```
